```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
```

```python
from lasts import Lasts
import numpy as np
import keras
from datasets import build_cbf
from joblib import load
from blackbox_wrapper import BlackboxWrapper
from lore_wrapper import LoreWrapper
```

```python
# IMPORT DATASET
random_state = 0
dataset_name = "cbf"

(X_train, y_train, X_val, y_val, 
 X_test, y_test, X_exp_train, y_exp_train, 
 X_exp_val, y_exp_val, X_exp_test, y_exp_test) = build_cbf(n_samples = 600, 
                                             random_state = random_state)
```

```python
# IMPORT BLACKBOX
knn = load(parentdir + "/trained_models/cbf/cbf_knn.joblib")
resnet = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_resnet.h5")

blackbox = resnet
```

```python
# WRAP BLACKBOX
blackbox = BlackboxWrapper(resnet, 3, 2)
```

```python
# IMPORT AUTOENCODER
ae = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_ae.h5")

autoencoder = ae
encoder = autoencoder.layers[1]
decoder = autoencoder.layers[2]
```

```python
# CHOOSE INSTANCE TO EXPLAIN
x = X_exp_test[1].ravel().reshape(1,-1,1)
```

```python
K = X_exp_test[1:]
neighborhood_generator = LoreWrapper(
    blackbox, 
    encoder, 
    decoder
    )
neighborhood_generator.fit(K)

lasts = Lasts(blackbox, 
              encoder, 
              decoder,
              x,
              neighborhood_generator
              )

#out = lasts.generate_neighborhood(verbose=True, mu=2)
out = lasts.generate_neighborhood(verbose=True, ngen = 1, size = 500, samples = 500)
lasts.plot_latent_space(K = encoder.predict(X_exp_train))

lasts.build_latent_tree()
lasts.plot_latent_space_latent_tree(K = encoder.predict(X_exp_train))


#lasts.morphing_matrix()
#lasts.plot_exemplars_and_counterexemplars()

"""
lasts.build_shapelet_explainer_latent_tree(shapelet_model_params={"max_iter":50}, 
                                           rebuild_tree = True)
lasts.shapelet_explainer_fidelity(latent_tree = True)"""
```
