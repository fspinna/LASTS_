```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
```

```python
from joblib import load
from blackbox_wrapper import BlackboxWrapper
import keras
from datasets import build_cbf
import time
from tests import build_multiple_lore_lasts
from lasts import load_multiple_lasts
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_name = "cbf"
blackbox_name = "knn"
file_path = parentdir + "/saved/" + dataset_name + "_" + blackbox_name + "_" + time.strftime("%Y%m%d_%H%M%S")
```

```python
random_state = 0
(X_train, y_train, X_val, y_val, 
 X_test, y_test, X_exp_train, y_exp_train, 
 X_exp_val, y_exp_val, X_exp_test, y_exp_test) = build_cbf(n_samples = 600, 
                                             random_state = random_state)
```

```python
knn = load(parentdir + "/trained_models/cbf/cbf_knn.joblib")
#resnet = keras.models.load_model("./trained_models/cbf/cbf_resnet.h5")
autoencoder = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_ae.h5")

blackbox = BlackboxWrapper(knn, 2, 1)
#blackbox = BlackboxWrapper(resnet, 3, 2)
encoder = autoencoder.layers[1]
decoder = autoencoder.layers[2]
```

```python
generator_params = {"verbose":True, 
                    "ngen":1, 
                    "size":50, 
                    "samples": 50}
shapelet_model_params = {"max_iter":2}
```

```python
%%time
lasts_list = build_multiple_lore_lasts(
    blackbox,
    encoder,
    decoder,
    X_exp_test,
    file_path,
    generator_params = generator_params,
    shapelet_model_params = shapelet_model_params,
    verbose = True,
    n_instances_to_explain = None
    )
```
