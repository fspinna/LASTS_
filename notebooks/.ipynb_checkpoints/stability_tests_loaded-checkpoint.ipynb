```python
from joblib import load
from utils import boxplot_from_df, baseline_error
import pandas as pd
from lasts import load_multiple_lasts
import numpy as np
from sklearn.neighbors import NearestNeighbors
from collections import defaultdict
from scipy.spatial.distance import cdist
from sklearn.metrics import mean_absolute_error
```

```python
from utils import boxplot_from_df
```

```python
dict_stab = load(parentdir + "/test_metrics/precal_predictions.joblib")
```

```python
lasts_cbf_knn = load_multiple_lasts(parentdir + "/saved/knn/cbf/matched_uniform_saxdt", simple_load=True)
lasts_cbf_resnet = load_multiple_lasts(parentdir + "/saved/resnet/cbf/matched_uniform_saxdt", simple_load=True)

lasts_coffee_knn = load_multiple_lasts(parentdir + "/saved/knn/coffee/matched_uniform_saxdt", simple_load=True)
lasts_coffee_resnet = load_multiple_lasts(parentdir + "/saved/resnet/coffee/matched_uniform_saxdt", simple_load=True)

lasts_ecg200_knn = load_multiple_lasts(parentdir + "/saved/knn/ecg200/matched_uniform_saxdt", simple_load=True)
lasts_ecg200_resnet = load_multiple_lasts(parentdir + "/saved/resnet/ecg200/matched_uniform_saxdt", simple_load=True)

lasts_gunpoint_knn = load_multiple_lasts(parentdir + "/saved/knn/gunpoint/matched_uniform_saxdt", simple_load=True)
lasts_gunpoint_resnet = load_multiple_lasts(parentdir + "/saved/resnet/gunpoint/matched_uniform_saxdt", simple_load=True)

lasts_lists = [lasts_cbf_knn, lasts_cbf_resnet, lasts_coffee_knn, lasts_coffee_resnet, lasts_ecg200_knn,
               lasts_ecg200_resnet, lasts_gunpoint_knn, lasts_gunpoint_resnet]
names = ["cbf_knn", "cbf_resnet", "coffee_knn", "coffee_resnet", "ecg200_knn", "ecg200_resnet",
         "gunpoint_knn", "gunpoint_resnet"]
```

```python
verbose=True
latent=False
compute_lasts=True
divide_by_baseline=True
lasts_kwargs=dict()
combination_name="prova"
debug=False
```

```python
instabilities = defaultdict(list)
for i, key in enumerate(dict_stab):
    combination = dict_stab[key]
    X = list()
    Z = list()
    for lasts_ in lasts_lists[i]:
        X.append(lasts_.x.ravel())
        Z.append(lasts_.z.ravel())
    X = np.array(X)[:, :, np.newaxis]
    Z = np.array(Z)
    print(X.shape, Z.shape)
    if latent:
        nbrs = NearestNeighbors(n_neighbors=2, algorithm='ball_tree').fit(Z)
        distances, idxs = nbrs.kneighbors(Z)
    else:
        nbrs = NearestNeighbors(n_neighbors=2, algorithm='ball_tree').fit(X[:, :, 0])
        distances, idxs = nbrs.kneighbors(X[:, :, 0])

    if compute_lasts:
        if verbose:
            print("Computing LASTS instability... ", sep="")
        for i, idx in enumerate(idxs):
            #print(i, idx)
            instability_lasts_ = instability_lasts(
                combination[i],
                combination[idx[-1]],  # the first value is the idx of the instance itself
                ignore_mismatch=True,
                divide_by_baseline=divide_by_baseline,
                debug=debug,
                **lasts_kwargs
            )
            if instability_lasts_ is not np.nan:
                instabilities[key].append(instability_lasts_)
        if verbose:
            print("Done!")
```

```python
columns = list(instabilities.keys())
```

```python
for i in range(len(columns)):
    columns[i] = columns[i]+"_shap"
```

```python
df = list()
for key in instabilities:
    df.append(instabilities[key])
```

```python
def instability_lasts(pred_importances1, pred_importances2, ignore_mismatch=False, debug=False, 
                      divide_by_baseline=True):
    errors = list()
    if (pred_importances1[-1].sum() == 0) and (pred_importances2[-1].sum() == 0):
            errors.append(0)
    elif (pred_importances1[-1].sum() == 0) or (pred_importances2[-1].sum() == 0):
        pass
        if not ignore_mismatch:
            baseline = baseline_error(pred_importances1[-1])
            if baseline == 0:
                errors.append(1)
            else:
                errors.append(mean_absolute_error(pred_importances1[-1, :], pred_importances2[-1, :]) / 
                              baseline_error(pred_importances1[-1]))
    else:
        baseline = baseline_error(pred_importances1[-1])
        if baseline == 0:
            errors.append(1)
        else:
            error = mean_absolute_error(pred_importances1[-1, :], pred_importances2[-1, :]) / baseline_error(pred_importances1[-1])
            errors.append(error)
    
    #print(errors)
    pred_importances1 = pred_importances1[:-1]
    pred_importances2 = pred_importances2[:-1]
    if (len(pred_importances1) == 0) and (len(pred_importances2) == 0):
        errors.append(0)
    else:
        if not ignore_mismatch:
            #  doesn't ignore if there is a different number of not-contained subsequences i.e. the surplus vectors will
            #  be matched with all 0's vectors
            mismatch = len(pred_importances1) - len(pred_importances2)
            if mismatch != 0:
                if mismatch > 0:
                    if len(pred_importances2) == 0:
                        pred_importances2 = np.zeros(shape=(abs(mismatch), pred_importances1.shape[1]))
                    else:
                        pred_importances2 = np.append(pred_importances2,
                                                      np.zeros(shape=(abs(mismatch), pred_importances1.shape[1])), axis=0)
                else:
                    if len(pred_importances1) == 0:
                        pred_importances1 = np.zeros(shape=(abs(mismatch), pred_importances1.shape[1]))
                    else:
                        pred_importances1 = np.append(pred_importances1,
                                                      np.zeros(shape=(abs(mismatch), pred_importances1.shape[1])), axis=0)
        else:
            if ((len(pred_importances1) == 0) or (len(pred_importances2) == 0)):
                return np.mean(errors)
        distance_matrix = cdist(pred_importances1, pred_importances2, metric=mean_absolute_error)
        #print(distance_matrix)
        if debug:
            print(distance_matrix)
        while (distance_matrix.shape[1] > 0) and (len(distance_matrix) > 0):
            if debug:
                print(distance_matrix)
            to_delete = np.unravel_index(np.argmin(distance_matrix), shape=distance_matrix.shape)
            if divide_by_baseline:
                baseline = baseline_error(pred_importances1[to_delete[0]])
                if baseline == 0:
                    errors.append(1)  # FIXME: limit case
                else:
                    errors.append(distance_matrix.min() / baseline)
            else:
                errors.append(distance_matrix.min())
            distance_matrix = np.delete(distance_matrix, to_delete[0], axis=0)
            distance_matrix = np.delete(distance_matrix, to_delete[1], axis=1)
    """ print("--")
    print(np.mean(errors))
    print("--")"""
    return np.mean(errors)
```
