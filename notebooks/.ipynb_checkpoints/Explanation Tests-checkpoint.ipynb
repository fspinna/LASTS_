```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
```

```python
from shap_explainer import get_latent_shap_explainer
import shap
from datasets import build_cbf
from sklearn.model_selection import train_test_split
from autoencoders import Autoencoder
from joblib import load
from blackbox_wrapper import BlackboxWrapper
from lore_wrapper import LoreWrapper
from lasts import Lasts
from lasts_old import LastsOld
import numpy as np
from growing_spheres import GrowingSpheres
import keras
from utils import reconstruction_accuracy
```

```python
# IMPORT DATASET
random_state = 0
dataset_name = "cbf"

(X_train, y_train, X_val, y_val, 
 X_test, y_test, X_exp_train, y_exp_train, 
 X_exp_val, y_exp_val, X_exp_test, y_exp_test) = build_cbf(n_samples = 600, 
                                             random_state = random_state)
```

```python
# IMPORT BLACKBOX
knn = load(parentdir + "/trained_models/cbf/cbf_knn.joblib")
resnet = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_resnet.h5")

blackbox = resnet
```

```python
# WRAP BLACKBOX
blackbox = BlackboxWrapper(resnet, 3, 2)
```

```python
# IMPORT AUTOENCODER
aae_normal = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_aae_normal.h5")

autoencoder = aae_normal
encoder = autoencoder.layers[1]
decoder = autoencoder.layers[2]
```

```python
# CHECK RECONSTRUCTION ACCURACY
print(reconstruction_accuracy(X_exp_test, encoder, decoder, blackbox))
```

```python
# CHOOSE INSTANCE TO EXPLAIN
x = X_exp_test[1].ravel().reshape(1,-1,1)
```

### GROWING SPHERES

```python
neighborhood_generator = GrowingSpheres(blackbox, decoder)   
lasts = Lasts(blackbox, 
              encoder, 
              decoder,
              x,
              neighborhood_generator
              )
```

```python
# GENERATE NEIGHBORHOOD
out = lasts.generate_neighborhood(verbose=True, mu=2)
```

```python
# VARIOUS PLOTS
lasts.plot_latent_space(scatter_matrix = False)
lasts.plot_latent_space(K = encoder.predict(X_exp_train))
lasts.morphing_matrix()
lasts.plot_exemplars_and_counterexemplars()
```

```python
# SHAPELET EXPLAINER
lasts.build_shapelet_explainer(shapelet_model_params={"max_iter":50}, binarize_labels = True)
```

```python
# SHAPELET TREE
lasts._sbgdt._graph
```

```python
# VARIOUS PLOTS
lasts.plot_shapelet_space()
lasts.plot_binary_heatmap()
lasts.plot_factual_and_counterfactual()
```

### LORE (no latent tree)

```python
K = X_exp_test[1:]
neighborhood_generator = LoreWrapper(
    blackbox, 
    encoder, 
    decoder
    )
neighborhood_generator.fit(K)

lasts = Lasts(blackbox, 
                  encoder, 
                  decoder,
                  x,
                  neighborhood_generator
                  )
```

```python
out = lasts.generate_neighborhood(verbose=True)
```

```python
# VARIOUS PLOTS
lasts.plot_latent_space(scatter_matrix = False)
lasts.plot_latent_space(K = encoder.predict(X_exp_test))
lasts.morphing_matrix()
lasts.plot_exemplars_and_counterexemplars()
```

```python
# SHAPELET EXPLAINER
lasts.build_shapelet_explainer(shapelet_model_params={"max_iter":50}, binarize_labels = True)
```

```python
# SHAPELET TREE
lasts._sbgdt._graph
```
