```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
from shap_explainer import get_latent_shap_explainer
import shap
```

```python
from pyts.datasets import make_cylinder_bell_funnel
from sklearn.model_selection import train_test_split
from autoencoders import Autoencoder
from joblib import load
from blackbox_wrapper import BlackboxWrapper
from lore_wrapper import LoreWrapper
from lasts import Lasts
import numpy as np
from growing_spheres import GrowingSpheres
from blackboxes import build_resnet

random_state = 0
dataset_name = "cbf"


X_all, y_all = make_cylinder_bell_funnel(n_samples = 600, random_state = random_state)
X_all = X_all[:,:,np.newaxis]

print("DATASET INFO:")
print("X SHAPE: ", X_all.shape)
print("y SHAPE: ", y_all.shape)
unique, counts = np.unique(y_all, return_counts=True)
print("\nCLASSES BALANCE")
for i, label in enumerate(unique):
    print(label, ": ", round(counts[i]/sum(counts), 2))

# BLACKBOX/EXPLANATION SETS SPLIT
X_train, X_exp, y_train, y_exp = train_test_split(
    X_all, 
    y_all, 
    test_size=0.3, 
    stratify = y_all, random_state=random_state
    )

# BLACKBOX TRAIN/TEST SETS SPLIT
X_train, X_test, y_train, y_test = train_test_split(
    X_train, 
    y_train, 
    test_size=0.2, 
    stratify = y_train, 
    random_state=random_state
    )

# BLACKBOX TRAIN/VALIDATION SETS SPLIT
X_train, X_val, y_train, y_val = train_test_split(
    X_train, 
    y_train, 
    test_size=0.2, 
    stratify = y_train, 
    random_state=random_state
    )

# EXPLANATION TRAIN/TEST SETS SPLIT
X_exp_train, X_exp_test, y_exp_train, y_exp_test = train_test_split(
    X_exp, 
    y_exp, 
    test_size=0.2, 
    stratify = y_exp, 
    random_state=random_state
    )

# EXPLANATION TRAIN/VALIDATION SETS SPLIT
X_exp_train, X_exp_val, y_exp_train, y_exp_val = train_test_split(
    X_exp_train, y_exp_train, 
    test_size=0.2, 
    stratify = y_exp_train, 
    random_state=random_state
    )

print("\nSHAPES:")
print("BLACKBOX TRAINING SET: ", X_train.shape)
print("BLACKBOX VALIDATION SET: ", X_val.shape)
print("BLACKBOX TEST SET: ", X_test.shape)
print("EXPLANATION TRAINING SET: ", X_exp_train.shape)
print("EXPLANATION VALIDATION SET: ", X_exp_val.shape)
print("EXPLANATION TEST SET: ", X_exp_test.shape)

n_timesteps, n_outputs, n_features = X_train.shape[1], len(np.unique(y_all)), 1 
print("\nTIMESTEPS: ", n_timesteps)
print("N. LABELS: ", n_outputs)

knn = load(parentdir + "/trained_models/cbf/cbf_knn.joblib")
resnet = keras.models.load_model(parentdir + "/trained_models/cbf/cbf_resnet.h5")

params = {"input_shape": (n_timesteps,1),
      "n_blocks": 8, 
      "latent_dim": 2,
      "encoder_latent_layer_type": "variational",
      "encoder_args": {"filters":[2,4,8,16,32,64,128,256], 
                        "kernel_size":[21,18,15,13,11,8,5,3], 
                        "padding":"same", 
                        "activation":"elu", 
                        "pooling":[1,1,1,1,1,1,1,1]}
     }

aut = Autoencoder(verbose = False, **params)
encoder, decoder, autoencoder = aut.build()
autoencoder.load_weights(parentdir + "/trained_models/cbf/cbf_vae.hdf5")

blackbox = BlackboxWrapper(knn, 2, 1)
blackbox = BlackboxWrapper(resnet, 3, 2)
encoder = autoencoder.layers[1]
decoder = autoencoder.layers[2]

x = X_exp_test[0].ravel().reshape(1,-1,1)

"""K = X_exp_test[1:]

neighborhood_generator = LoreWrapper(
    blackbox, 
    encoder, 
    decoder
    )
neighborhood_generator.fit(K)"""
neighborhood_generator = GrowingSpheres(blackbox, decoder)
lasts = Lasts(blackbox, 
              encoder, 
              decoder,
              x,
              neighborhood_generator
              )

out = lasts.generate_neighborhood(verbose=True, mu=2)
```

### INTERACTIVE LATENT SPACE

```python
import matplotlib.pyplot as plt
z = lasts.z
z_tilde = lasts.z_tilde
z_label = lasts.x_label
Z = lasts._Z
y = lasts._y
from scipy.stats import norm
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
def dec(**kwargs):
    z = np.array(list(kwargs.values())).reshape(1,-1)
    z_tilde = decoder.predict(z)
    z_label = blackbox.predict(z_tilde)[0]
    z_color = "#2ca02c" if z_label == y_exp_test[0] else "#d62728"
    z_tilde = z_tilde.ravel()
    plt.figure(figsize=(10,3))
    plt.title("label " + str(z_label))
    plt.plot(z_tilde, c = z_color)
    plt.gca().set_ylim((-1,8))
    plt.show()
    
def dec2(**kwargs):
    z = np.array(list(kwargs.values())).reshape(1,-1)
    z_tilde = decoder.predict(z)
    x_label = blackbox.predict(z_tilde)[0]
    z_color = "#2ca02c" if x_label == y_exp_test[0] else "#d62728"
    z_tilde = z_tilde.ravel()
    
    fig, axs = plt.subplots(1,2, figsize = (15,3), gridspec_kw={'width_ratios': [3, 1]})
    #axs[0].title("label " + str(z_label))
    axs[0].plot(z_tilde, c = z_color)
    axs[0].set_ylim((-1,8))
    
    exemplars = np.argwhere(y == z_label)
    counterexemplars = np.argwhere(y != z_label)
    
    axs[1].scatter(
        Z[:,0][exemplars], 
        Z[:,1][exemplars], 
        c = "#2ca02c", 
        alpha = 0.5
        )
    axs[1].scatter(
        Z[:,0][counterexemplars], 
        Z[:,1][counterexemplars], 
        c = "#d62728", 
        alpha = 0.5
        )
    
    # marks the instance to explain with an X
    axs[1].scatter(
        z.ravel()[0], 
        z.ravel()[1],
        c = "royalblue", 
        marker = "X", 
        edgecolors = "white",
        s = 400) 
    axs[1].set_xlim(-2.5, 0.5)
    axs[1].set_ylim((-1.5,1.5))
    plt.show()
```

```python
%matplotlib inline
minimum = -5
maximum = 5
variables = dict.fromkeys([str(i) for i in range(len(z.ravel()))])
for key in variables.keys():
    variables[key] = widgets.FloatSlider(value=0,
                                         min=minimum,
                                         max=maximum,
                                         step=0.05,
                                         description = "feature " + key)
interact(
    dec2, 
    **variables
)
```
