```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
from ts_syege import generate_synthetic_ts_classifier_s_dataset, get_subsequence_explanation
from ts_syege import generate_synthetic_ts_classifier_b_dataset, get_boundary_explanation
```

```python
from pyts.datasets import load_gunpoint, load_coffee
```

```python
X_train, X_test, y_train, y_test = load_gunpoint(return_X_y=True)

m = X_train.shape[1]

k_min = 4  # min nbr of subsequences
k_max = 8  # max nbr of subsequences
l_min = 4  # min length
l_max = 16  # max length
p = 50  # thr closeness subsequences
random_state = 0

stc_s = generate_synthetic_ts_classifier_s_dataset(X_train, y_train, m, k_min, k_max, l_min, l_max, p, random_state)

predict = stc_s['predict']
predict_proba = stc_s['predict_proba']
subsequences = stc_s['subsequences']

y = predict(X_train)
print(np.unique(y, return_counts=True))
```

```python
idx = 12
x = X_train[idx]
#print(x)
#print(y[idx])
plt.title('b(x) = %s' % y[idx])
plt.plot(x, c='k')
plt.show()

exp = get_subsequence_explanation(x, subsequences)
print(exp)

plt.title('b(x) = %s' % y[idx])
plt.plot(x, c='k')
for e in exp:
    s, idx = e
    plt.plot(range(idx, idx + len(s)), s, c='g', lw=3)
plt.show()
```

```python
sts_b = generate_synthetic_ts_classifier_b_dataset(X_train, y_train, m, k_min, k_max, l_min, l_max, random_state)

predict = sts_b['predict']
predict_proba = sts_b['predict_proba']
boundaries = sts_b['boundaries']

y = predict(X_train)
print(np.unique(y, return_counts=True))
```

```python
idx = 0
x = X_train[idx]
#print(x)
#print(y[idx])
plt.title('b(x) = %s' % y[idx])
plt.plot(x, c='k')
plt.show()

exp = get_boundary_explanation(x, boundaries)
print(exp)

plt.title('b(x) = %s' % y[idx])
plt.plot(x, c='k')
all_w = [e[2] for e in exp]
tot_w = np.sum(all_w)
for e in exp:
    s, l, w = e
    plt.axvspan(s, s + l, alpha=w/tot_w, color='g')
plt.show()
```
