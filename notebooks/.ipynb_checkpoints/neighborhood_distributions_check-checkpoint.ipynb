```python
from neighborhood_generators import vicinity_sampling, check_neighborhood_norm
from utils import euclidean_norm, plot_norm_distributions
import matplotlib.pyplot as plt
import numpy as np
```

##### if we extract randomly (random normal) n point and we generate a neighborhood around them, does this neighborhood still have a random distribution? Are the euclidean norm similar to that of a random distribution?

```python
size=(1,2)
methods = ["gaussian_global", "gaussian", "gaussian_matched", "uniform_sphere", "uniform_sphere_scaled"]
n = 1000
n_single_neighborhood=1
```

##### se il threshold è alto la differenza di distribuzione è evidente

```python
threshold=4
Ts = []
for method in methods:
    Z, T = check_neighborhood_norm(
        size, method, n=n, threshold=threshold, n_single_neighborhood=n_single_neighborhood
    )
    Ts.append(T)
    plt.scatter(Z[:,0], Z[:,1], alpha=.1)
    plt.title(method)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    
Z = np.random.normal(size=(n, size[1]))
Ts.append(euclidean_norm(Z))
methods.append("true_normal")
plot_norm_distributions(Ts, labels=methods)
methods.remove("true_normal")
```

##### se è basso il distibution mismatch è quasi inesistente (perchè i punti estratti sono molto vicini a quello estratto con una normale)

```python
threshold=0.5
Ts = []
for method in methods:
    Z, T = check_neighborhood_norm(
        size, method, n=n, threshold=threshold, n_single_neighborhood=n_single_neighborhood
    )
    Ts.append(T)
    plt.scatter(Z[:,0], Z[:,1], alpha=.1)
    plt.title(method)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    
Z = np.random.normal(size=(n, size[1]))
Ts.append(euclidean_norm(Z))
methods.append("true_normal")
plot_norm_distributions(Ts, labels=methods)
methods.remove("true_normal")
```

##### lo stesso vale in spazi multidimensionali (a parità di threshold il distribution mismatch è più evidente)

```python
size=(1,100)
threshold=0.5
Ts = []
for method in methods:
    Z, T = check_neighborhood_norm(
        size, method, n=n, threshold=threshold, n_single_neighborhood=n_single_neighborhood
    )
    Ts.append(T)
    
Z = np.random.normal(size=(n, size[1]))
Ts.append(euclidean_norm(Z))
methods.append("true_normal")
plot_norm_distributions(Ts, labels=methods)
methods.remove("true_normal")

threshold=4
Ts = []
for method in methods:
    Z, T = check_neighborhood_norm(
        size, method, n=n, threshold=threshold, n_single_neighborhood=n_single_neighborhood
    )
    Ts.append(T)
    
Z = np.random.normal(size=(n, size[1]))
Ts.append(euclidean_norm(Z))
methods.append("true_normal")
plot_norm_distributions(Ts, labels=methods)
methods.remove("true_normal")
```
