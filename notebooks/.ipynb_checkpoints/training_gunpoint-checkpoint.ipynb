### Blackbox

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, mean_squared_error, accuracy_score
from sklearn.model_selection import StratifiedKFold, RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
#metriche usate nella grid search
scoring = {'MSE':make_scorer(mean_squared_error, greater_is_better=False), 'Accuracy': make_scorer(accuracy_score)}

#iperparametri
n_neighbors = list(range(1, 20))
metric = ["euclidean"]
#weights = ['uniform', 'distance']
weights = ['distance']
param_grid = dict(n_neighbors = n_neighbors, weights = weights, metric = metric)

#Definition of algorithm and GridSearch
knn = KNeighborsClassifier(algorithm='auto')
cv = StratifiedKFold(random_state=random_state, n_splits=3)
grid = GridSearchCV(knn, param_grid = param_grid, cv = cv, scoring = scoring, n_jobs=-1, 
                    return_train_score=True, refit='MSE', verbose=1)
```

```python
%%time
grid.fit(X_train[:,:,0], y_train)
results = grid.cv_results_
```

```python
print(grid.best_params_)
df_results = pd.DataFrame(results)
df_results.at[grid.best_index_, 'mean_test_Accuracy']
```

```python
knn = KNeighborsClassifier(**grid.best_params_)
```

```python
knn.fit(X_train.reshape(X_train.shape[:2]), y_train)
```

```python
from lasts import choose_z
from sklearn.metrics import accuracy_score
def reconstruction_accuracy_vae(X, encoder, decoder, blackbox, repeat=1, n=100, verbose=True):
    y = blackbox.predict(X)
    accuracies = []
    for i in range(repeat):
        Z = list()
        for x in X:
            z = choose_z(x[np.newaxis,:,:], encoder, decoder, n, verbose)
            Z.append(z.ravel())
        Z = np.array(Z)
        y_tilde = blackbox.predict(decoder.predict(Z))
        accuracy = accuracy_score(y, y_tilde)
        accuracies.append(accuracy)
    accuracies = np.array(accuracies)
    accuracies_mean = accuracies.ravel().mean()
    accuracies_std = np.std(accuracies.ravel())
    if verbose:
        print("Accuracy:", accuracies_mean, "", accuracies_std)
    return accuracies_mean
```

```python
reconstruction_accuracy_vae(X_exp_test, encoder, decoder, blackbox, repeat=1)
```

```python
def plot_reconstruction_vae(X, encoder, decoder, figsize=(20, 15), n=0):
    Z = list()
    for x in X:
        z = choose_z(x[np.newaxis, :, :], encoder, decoder)
        Z.append(z.ravel())
    Z = np.array(Z)
    X_tilde = decoder.predict(Z)
    g = 1
    plt.figure(figsize=figsize)
    for i in range(n, n + 5):
        # display original
        ax = plt.subplot(5, 1, g)
        g += 1
        plt.plot(X[i], label="real")
        plt.plot(X_tilde[i], label="reconstructed")
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.legend()
    plt.show()
```
