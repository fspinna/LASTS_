```python
from shap_explainer import get_latent_shap_explainer, get_latent_shap_explainer_binary
import shap
from datasets import build_gunpoint
from sklearn.model_selection import train_test_split
from joblib import load
from blackbox_wrapper import BlackboxWrapper
from lasts import Lasts
import numpy as np
import keras
from utils import reconstruction_accuracy, plot_labeled_latent_space_matrix
from variational_autoencoder import load_model
from neighborhood_generators import NeighborhoodGenerator
from utils import compute_norm, choose_z
```

```python
# IMPORT DATASET
random_state = 0
dataset_name = "gunpoint"

(X_train, y_train, X_val, y_val, 
 X_test, y_test, X_exp_train, y_exp_train, 
 X_exp_val, y_exp_val, X_exp_test, y_exp_test) = build_gunpoint(random_state = random_state)
```

```python
# IMPORT BLACKBOX
resnet = keras.models.load_model(parentdir + "/trained_models/gunpoint/gunpoint_resnet.h5")
```

```python
# WRAP BLACKBOX
blackbox = BlackboxWrapper(resnet, 3, 2)
```

```python
# IMPORT AUTOENCODER
_, _, autoencoder = load_model(parentdir + "/trained_models/gunpoint/gunpoint_vae")

encoder = autoencoder.layers[2]
decoder = autoencoder.layers[3]
```

```python
# CHECK RECONSTRUCTION ACCURACY
reconstruction_accuracy(X_exp_test, encoder, decoder, blackbox)
```

```python
# CHOOSE INSTANCE TO EXPLAIN
x = X_exp_test[2].ravel().reshape(1,-1,1)
z = choose_z(x, encoder, decoder)
```

```python
K = encoder.predict(X_exp_train)
```

```python
plot_labeled_latent_space_matrix(K, blackbox.predict(decoder.predict(K)))
```

### GENERATORS

```python
neighborhood_generator = NeighborhoodGenerator(blackbox, decoder)
neigh_kwargs = {
        "balance": False,
        "n": 500,
        "n_search": 10000,
        "threshold": 2,
        "kind": "gaussian",
        "distribution": np.random.normal,
        "distribution_kwargs": dict(),
        "fine_search": True,
        "verbose": True,
        "stopping_ratio": 0.1
    }

lasts_ = Lasts(blackbox, encoder, decoder, x, neighborhood_generator, z_fixed=z)
```

```python
# GENERATE NEIGHBORHOOD
out = lasts_.generate_neighborhood(**neigh_kwargs)
```

```python
import matplotlib.pyplot as plt
z = lasts_.z
z_tilde = lasts_.z_tilde
z_label_fixed = lasts_.x_label
Z = lasts_._Z
y = lasts_._y
K = encoder.predict(X_exp_train)
from scipy.stats import norm
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
from lasts import plot_latent_space_matrix
```

```python
from lasts import plot_latent_space_matrix
def dec3(**kwargs):
    z = np.array(list(kwargs.values())).reshape(1,-1)
    z_tilde = decoder.predict(z)
    x_label = blackbox.predict(z_tilde)[0]
    z_color = "#2ca02c" if x_label == z_label_fixed else "#d62728"
    z_tilde = z_tilde.ravel()
    
    plt.figure(figsize=(20,3))
    plt.plot(z_tilde, c = z_color)
    plt.gca().set_ylim((z_tilde.min() - 0.5,
                     z_tilde.max() + 0.5 ))
    plt.show()
    
    plot_latent_space_matrix(Z, y, z, z_label_fixed)
```

```python
def dec2(**kwargs):
    z = np.array(list(kwargs.values())).reshape(1,-1)
    z_tilde = decoder.predict(z)
    x_label = blackbox.predict(z_tilde)[0]
    z_color = "#2ca02c" if x_label == z_label_fixed else "#d62728"
    z_tilde = z_tilde.ravel()
    
    fig, axs = plt.subplots(2,1, figsize = (20,10), gridspec_kw={'height_ratios': [1, 5]})
    #axs[0].title("label " + str(z_label))
    axs[0].plot(z_tilde, c = z_color)
    axs[0].set_ylim((lasts_._Z_tilde.min() - 0.5,
                     lasts_._Z_tilde.max() + 0.5 ))
    
    exemplars = np.argwhere(y == z_label_fixed)
    counterexemplars = np.argwhere(y != z_label_fixed)
    
    axs[1].scatter(
        K[:,0], 
        K[:,1], 
        c = "gray", 
        alpha = 0.5
        )
    
    axs[1].scatter(
        Z[:,0][exemplars], 
        Z[:,1][exemplars], 
        c = "#2ca02c", 
        alpha = 0.5
        )
    axs[1].scatter(
        Z[:,0][counterexemplars], 
        Z[:,1][counterexemplars], 
        c = "#d62728", 
        alpha = 0.5
        )
    
    # marks the instance to explain with an X
    axs[1].scatter(
        z.ravel()[0], 
        z.ravel()[1],
        c = "royalblue", 
        marker = "X", 
        edgecolors = "white",
        s = 400) 
    #axs[1].set_xlim(-1.5, 1.5)
    #axs[1].set_ylim((-1.5,1.5))
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
%matplotlib inline
mins = lasts_._Z.min(axis=0)
maxs = lasts_._Z.max(axis=0)

variables = dict.fromkeys([str(i) for i in range(len(z.ravel()))])
for i, key in enumerate(variables.keys()):
    variables[key] = widgets.FloatSlider(value=(mins[i]+maxs[i])/2,
                                         min=mins[i],
                                         max=maxs[i],
                                         step=0.01,
                                         description = "feature " + key)
interact(
    dec2, 
    **variables
)
```
