```python
import os,sys,inspect
currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0,parentdir)
```

```python
import pandas as pd
from joblib import load
from utils import boxplot_from_df
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_normal = pd.read_csv(parentdir + "/test_metrics/stability_manifest.csv", sep=";")
df = load(parentdir + "/test_metrics/stability_manifest_boxplot_ready.joblib")
```

```python
figsize=(20,3)
dpi=300
ylabel="instability"
xticks=["lasts", "shap"]*8
fontsize=20
```

```python
x_labels = ['cbf-knn', 'cbf-res', 'cof-knn','cof-res', 'ecg-knn', 'ecg-res', 'gun-knn',
            'gun-res']
```

```python
medianprops = dict(linestyle='-', linewidth=3, color='#8c510a')
meanprops = dict(marker='D', markeredgecolor='#003c30', markerfacecolor='#003c30')
boxprops = dict(linestyle='-', linewidth=3, color='#01665e')
whiskerprops = dict(linestyle='-', linewidth=3, color='#01665e')
capprops = dict(linestyle='-', linewidth=3, color='#01665e')
plt.figure(figsize=(figsize), dpi=dpi)
plt.boxplot(df, medianprops=medianprops, boxprops=boxprops, capprops=capprops, whiskerprops=whiskerprops,
            meanprops=meanprops, showmeans=False, showfliers=False)

#plt.axhline(y=1, color='r', linestyle='--')
plt.xticks(np.array(list(range(len(df_normal.columns)))) + 1, xticks)
plt.ylabel(ylabel, fontsize=fontsize)
x_pos = np.array(list(range(len(df_normal.columns)))) + 1.5
for i in range(len(df_normal.columns)):
    if i%2 == 0:
        plt.text(x_pos[i], plt.gca().get_ylim()[0] - 0.8, x_labels[int(i/2)], fontsize=fontsize, 
                 horizontalalignment="center")
    else:
        plt.axvline(x=x_pos[i], color='black', linestyle='-')
plt.tick_params(axis='both', which='major', labelsize=fontsize)
plt.show()
```
